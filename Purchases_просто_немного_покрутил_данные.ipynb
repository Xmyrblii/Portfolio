```python
retail = pd.read_csv('C:/Users/Admin/Desktop/karpov/data.csv', encoding='Windows-1251')
```

```python
retail = retail.drop_duplicates()
```

```python
aboba = retail.query('Country == "Germany"')\
      .groupby('CustomerID', as_index = False)\
      .agg({'InvoiceNo' : 'nunique'})
```

```python
N = aboba.InvoiceNo.quantile(0.8)
```

```python
keka = aboba.query('InvoiceNo > @N')
```

```python
keka
```

```python
top_germany = keka.CustomerID
```

```python
top_retail_germany = retail.loc[retail.CustomerID.isin(top_germany)]
```

```python
top_retail_germany
```

```python
top_retail_germany.groupby('StockCode', as_index = False)\
                  .agg({'InvoiceNo' : 'count'})\
                  .sort_values(by = 'InvoiceNo' )
```

```python
retail = retail.assign(Revenue = retail.Quantity * retail.UnitPrice)
```

```python
retail
```

```python
top_revenue = retail.groupby('InvoiceNo').agg(TotalRevenue = ('Revenue', 'sum'))\
                                         .sort_values('TotalRevenue', ascending = False)
```
