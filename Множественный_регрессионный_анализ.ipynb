```python
import pandas as pd
```

```python
import statsmodels.api as sm

import statsmodels.formula.api as smf

import numpy as np
```

```python
df = pd.read_csv('cars.csv')
```

```python
df
```

```python
df.info()
```

```python
df['company'] = df.CarName.apply(lambda x: x.split(' ')[0])
```

```python
df.company
```

```python
df.company.nunique()
```

```python
df = df.drop(columns=['CarName', 'car_ID']) 
```

```python
df
```

```python
df.company = df.company.str.lower().replace('maxda', 'mazda') 
```

```python
df.company = df.company.replace('porcshce', 'porsche') 
```

```python
df.company = df.company.replace('toyouta', 'toyota') 
```

```python
df.company = df.company.replace('vokswagen', 'volkswagen') 
```

```python
df.company = df.company.replace('vw', 'volkswagen') 
```

```python
df.company.nunique()
```

```python
df
```

```python
df1 = pd.DataFrame(df, columns=['price', 'company', 'fueltype', 'aspiration','carbody', 'drivewheel', 'wheelbase', 'carlength','carwidth', 'curbweight', 'enginetype', 'cylindernumber', 'enginesize', 'boreratio','horsepower']) 
```

```python
df1
```

```python
df1.corr()
```

```python
df1.dtypes
```

```python
df2 = pd.DataFrame(df1, columns=['price', 'wheelbase', 'carlength','carwidth', 'curbweight', 'enginesize', 'boreratio','horsepower']) 
```

```python
df2
```

```python
df1_dummy = pd.get_dummies(data=df1[['company', 'fueltype', 'aspiration', 'carbody', 'drivewheel', 'enginetype', 'cylindernumber' ]], drop_first = True)
```

```python
df1_dummy
```

```python
df3 = pd.concat([df1_dummy, df2], axis=1)
```

```python
df3
```

```python
mod = smf.ols(formula='price ~ horsepower', data=df3)
res = mod.fit()
print(res.summary())
```

```python
mod = smf.ols(formula='price ~ ['company_audi', 'company_bmw', 'company_buick', 'company_chevrolet',
       'company_dodge', 'company_honda', 'company_isuzu', 'company_jaguar',
       'company_mazda', 'company_mercury', 'company_mitsubishi',
       'company_nissan', 'company_peugeot', 'company_plymouth',
       'company_porsche', 'company_renault', 'company_saab', 'company_subaru',
       'company_toyota', 'company_volkswagen', 'company_volvo', 'fueltype_gas',
       'aspiration_turbo', 'carbody_hardtop', 'carbody_hatchback',
       'carbody_sedan', 'carbody_wagon', 'drivewheel_fwd', 'drivewheel_rwd',
       'enginetype_dohcv', 'enginetype_l', 'enginetype_ohc', 'enginetype_ohcf',
       'enginetype_ohcv', 'enginetype_rotor', 'cylindernumber_five',
       'cylindernumber_four', 'cylindernumber_six', 'cylindernumber_three',
       'cylindernumber_twelve', 'cylindernumber_two', 'price', 'wheelbase',
       'carlength', 'carwidth', 'curbweight', 'enginesize', 'boreratio',
       'horsepower'], data=df3)
res = mod.fit()
print(res.summary())
```
